```python
import pandas as pd
import math as m
import numpy as np
import scipy as sp 
import plotly.express as px
import plotly.graph_objects as go

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

```

```python
#ОСЬ X
data = {
    "time": [0, 0.5, 1, 1.6, 2.1, 2.6, 3.1, 3.4, 3.6, 3.9, 4.2, 4.7, 4.95, 5.24, 5.5, 5.75, 6, 6.28],
    "amplitude": [0, 0.5, 0.9, 0.98, 0.8, 0.4, -0.8, 1.6, -1.4, -1.6, 1.0, -2.0, 1.1, -1.1, -2.6, 1.8, 0, -1.05]
}

otr = np.arange(-25,76)
```

```python
#Сигнал задан функцией
def signal(time):
    A = 1
    T = 50
    fi = 0
    omega = (2*m.pi)/T
    f = A * m.sin(omega*time+fi)  #функция сигнала
    return f
```

```python
def wavelet(a, tau):
    f = lambda time : (1/a**0.5) * m.exp(-0.5*((time-tau)/a)**2)* (((time-tau)/a)**2-1)*signal(time)
    wave =  sp.integrate.quad(f, -256, 257)
    return round(wave[0],3)

```

```python
df_signal = pd.DataFrame({
    "Y - Амплитуда": [signal(t) for t in np.arange(-100, 101, 1)],
    "X - Время": [t for t in np.arange(-100, 101, 1)],
    "Z": [[wavelet(x, y) for x in np.arange(-100, 101, 1)] for y in np.arange(-100, 101, 1)] 
})

signal_chart = px.line(df_signal, 
                       x="X - Время", y="Y - Амплитуда", 
                       title='Гармоническое колебание')
signal_chart.show()
```

```python
fig = go.Figure(data=[go.Surface(z=df_signal['Z'])])

fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                                  highlightcolor="limegreen", project_z=True))
fig.update_layout(title='Mt Bruno Elevation')

fig.show()
```
